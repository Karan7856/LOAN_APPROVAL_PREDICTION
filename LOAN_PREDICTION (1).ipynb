```python
#Gradient Descent Demo (for single feature)
X_demo = np.array([1,2,3,4,5])
Y_demo = np.array([2,4,6,8,10])
m = b = 0
L = 0.01
epochs = 1000
n = len(X_demo)
for _ in range(epochs):
    Y_pred_demo = m * X_demo + b
    D_m = (-2/n) * sum(X_demo * (Y_demo - Y_pred_demo))
    D_b = (-2/n) * sum(Y_demo - Y_pred_demo)
    m -= L * D_m
    b -= L * D_b
print("Gradient Descent result: m=", m, ", b=", b)
```
